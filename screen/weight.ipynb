```python
import numpy as np
import pandas as pd
```

```python
import numpy as np

def generate_grid_points(step):
    result = []
    # 生成均匀分布的网格点
    values = np.linspace(0, 1, int(1//step))
    for value1 in values:
        values2 = np.linspace(0, 1-value1, int((1-value1)/step))
        for value2 in values2:
            result.append([value1, value2, 1-value1-value2])    
    return result
```

```python
# 每个轴上生成10个点，总共生成1000个点

grid_points = generate_grid_points(0.005)
random_points_color = np.array(grid_points)*255
df = pd.DataFrame()
df[['d', 'm', 'p']] = grid_points
df[['r','g','b']] = random_points_color
df.to_csv('random_weight.csv')
```

```python
weights = [[0,0.2,0.8],
           [0.2,0,0.8],
           [0,0.8,0.2],
           [0.2,0.8,0],
           [0.8,0.2,0],
           [0.8,0,0.2],   
           [0,0.4,0.6],
           [0,0.6,0.4],
           [0.4,0,0.6],
           [0.4,0.6,0],
           [0.6,0,0.4],
           [0.6,0.4,0],
           [0.6,0.2,0.2],
           [0.2,0.2,0.6],
           [0.2,0.6,0.2],
           [0.4,0.2,0.4],
           [0.2,0.4,0.4],
           [0.4,0.4,0.2],
           ]

df = pd.DataFrame()
df[['d', 'm', 'p']] = weights

colors = []
for weight in weights:
    tmp = np.array(weight)*255
    colors.append(tmp)
df[['r','g','b']] = colors

df.to_csv('weight.csv')
```

```python
df = pd.read_csv('pre_result.csv')
```

```python
pre_df = df[['Tm', 'Tp', 'Td']]
df[['max_min_tm', 'max_min_tp', 'max_min_td']] = (pre_df - pre_df.min()) /(pre_df.max()-pre_df.min())
```

```python
for i in range(len(weights)):
    weight = weights[i]
    df['score_%s'%i] = weight[0]*df['max_min_td'] - weight[1]*df['max_min_tm']-weight[2]*df['max_min_tp']
```

```python
size = 1000
result_df = pd.DataFrame()
for i in range(len(weights)):
    score_name = 'score_%s'%i
    df_ = df.sort_values(by=[score_name],ascending=[False]).head(size)
    df_.to_csv('pre_result_%s.csv'%score_name)
    new_df = df_[['Tm', 'Tp', 'Td']]
    new_df['score_id'] = [i] * size
    new_df['score'] = df_['score_%s'%i]
    result_df = pd.concat([result_df, new_df], axis=0)
```

```python
result_df.to_csv('duplicate_result.csv')
```

```python
result_df = result_df.drop_duplicates(subset=['Tm', 'Tp', 'Td'], keep='last')
result_df.to_csv('result.csv')
```
